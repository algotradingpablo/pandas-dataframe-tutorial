## Suma, resta, multiplicación, división

```python
df = pd.DataFrame({"Empleado": ["Rafael", "Sergio", "David", "Javier"],
                   "Oficio": ["Cientifico de datos", "Desarrollador backend", "Analista de ciberseguridad", "Ingeniero de redes"],
                   "Horas Semana 1": [36, 39, 38, 34], 
                   "Horas Semana 2": [40, 32, 36, 32], 
                   "Horas Semana 3": [34, 40, 34, 38], 
                   "Horas Semana 4": [36, 34, 40, 36],
                   "Salario Bruto": [3700, 3500, 3300, 3100],
                   "Porcentaje de impuestos": [35, 33, 31, 29]})
```

```python
df
```

```python
df["Horas totales trabajadas"] = df["Horas Semana 1"] + df["Horas Semana 2"] + df["Horas Semana 3"] + df["Horas Semana 4"]
```

```python
df
```

```python
df["Salario Bruto"] = df["Salario Bruto"] - 100
```

```python
df
```

```python
df["Salario Bruto"] = df["Salario Bruto"] + 100
```

```python
df
```

```python
df["Salario Bruto"] = df["Salario Bruto"] * 2
```

```python
df
```

```python
df["Salario Bruto"] = df["Salario Bruto"] // 2
```

```python
df
```

```python
df["Salario por hora trabajada"] = df["Salario Bruto"] / df["Horas totales trabajadas"]
```

```python
df
```

```python
df["Salario Neto"] = (1 - df["Porcentaje de impuestos"]/100) * df["Salario Bruto"]
```

```python
1 - df["Porcentaje de impuestos"]/100
```

```python
df
```

## Operadores: mayor, menor.

```python
df["Cobra más de 2250 netos"] = df["Salario Neto"] > 2250
```

```python
df
```

```python
df["Cobra menos de 23/hora"] = df["Salario por hora trabajada"] < 23
```

```python
df
```
